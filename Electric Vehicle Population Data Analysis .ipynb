# Data Cleaning

```python
df.head()
```

```python
df.dtypes
```

```python
for i in df:
    if df[i].dtypes=="0":
        df[i] = df[i].fillna(df[i].mode()[0])
    elif df[i].dtypes=="int64":
        df[i]= df[i].fillna(df[i].mean())
    elif df[i].dtypes=="float64":
        df[i]= df[i].fillna(df[i].mean())
```

```python
df.isnull().sum()
```

# Electric Vehicle type distribution

```python
plt.figure(figsize=(20,10))

# Count Plot
plt.subplot(1, 2, 1)
plt.title("Electric Vehicle Type Distribution")
sns.countplot(x='Electric Vehicle Type', data=df)

# Pie Chart
plt.subplot(1, 2, 2)
plt.title("Electric Vehicle Type Distribution %")
plt.pie(
    df['Electric Vehicle Type'].value_counts(), 
    labels=df['Electric Vehicle Type'].unique(), 
    autopct='%0.0f%%'
);
```

# State that have more concentration on EV

```python
ev=df['State'].value_counts().reset_index() 
ev.columns = ['State', 'Count']
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.title("State with highest EV concentration")
ax=sns.barplot(data=ev[ev['State'] =='WA'],x='State',y='Count')
plt.xlabel('State')
plt.ylabel('Number of Vehicles')

top_5= ev[(ev['State']!="WA")].nlargest(10,'Count')
plt.subplot(1,2,2)
sns.barplot(data=top_5, x="State",y='Count')
plt.title("State with highest EV concentration ['excluding WA']")
plt.xlabel('State')
plt.ylabel('Number of Vehicles');
```

# Top 10 country that have more concentration on EV

```python
plt.figure(figsize=(20,10))
ev=df["County"].value_counts().reset_index()
ev.columns = ["County" , "Count"]
ev =ev.nlargest(10,"Count")
plt.title("County with highest EV concentration")
plt.xlabel("County")
plt.ylabel("Number of Vehicles")
sns.barplot(data=ev,x="County",y="Count",palette='Blues_d');
```

# Top 10 Car Models

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
ev=df['Model'].value_counts().reset_index()
ev.columns=['Model','Count']
plt.title("Top 10 Model of ev")
plt.xlabel('Model')
plt.ylabel('Number of Vehicles');
ev=ev.nlargest(10,"Count")
sns.barplot(data=ev, x="Model",y='Count', palette='terrain');

plt.subplot(1,2,2)
ev=ev.nlargest(10,"Count")
plt.title("Top 10 Model of ev %")
plt.pie(ev.Count, labels=ev.Model.unique(), autopct="%0.0f%%");
```

# Year-on-Year Growth Plot

```python
annual_count = df.groupby('Model Year').size().reset_index(name='Vehicle Count')
annual_count = annual_count.sort_values(by='Model Year')
annual_count['Year-on-Year Growth (%)'] = annual_count['Vehicle Count'].pct_change() * 100

plt.figure(figsize=(10, 6))
ax = sns.lineplot(data=annual_count, x='Model Year', y='Vehicle Count', marker='o',color='darkorange')
ax.set_title("Year-on-Year Electric Vehicle Growth")
ax.set_xlabel("Model Year")
ax.set_ylabel("Vehicle Count")
ax.grid(True)
plt.show()
```
